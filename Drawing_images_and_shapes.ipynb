```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title='Image',image=None,size=8):
  w=image.shape[0]
  h=image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(aspect_ratio*size,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
#Making black square canvas
image=np.zeros((512,512,3),np.uint8)
image_gray=np.zeros((512,512),np.uint8)
imshow('black canvas-RGB Color',image,size=6)
imshow('black canvas-Grayscale Color',image_gray,size=6)
```

```python
#Drawing a line over a black canvas diagonally
cv2.line(image,(0,0),(511,511),(255,127,0), 5)
imshow('Black canvas with diagonal line',image,size=6)
```

```python
image=np.zeros((512,512,3),np.uint8)
cv2.rectangle(image,(100,100),(300,250),(127,50,127), 10)
imshow('Black canvas with reactangle',image,size=6)
```

```python
image=np.zeros((512,512,3),np.uint8)
cv2.circle(image,(350,350), 100, (15,150,50), -1)
imshow('Black canvas with green circles',image)
```

```python
image=np.zeros((512,512,3),np.uint8)
points=np.array([[10,50],[400,50],[90,200],[50,500]],np.int32)
points=points.reshape((-1,1,2))
cv2.polylines(image,[points],True,(0,0,255),3)
imshow('Black canvas with polygons',image,size=6)
```

```python
image=np.zeros((1000,1000,3),np.uint8)
ourString='Hello world!'
cv2.putText(image,ourString,(155,290),cv2.FONT_HERSHEY_COMPLEX,3,(40,200,0),4)
imshow('Messing with some text',image,size=6)
```
