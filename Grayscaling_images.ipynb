```python
def imshow(title='Image',image=None,size=10):
  w=image.shape[0]
  h=image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
image=cv2.imread('/content/images/Volleyball.jpeg')
imshow('Volleyball',image)
```

```python
imshow('Volleyball',image,size=6)
```

```python
#for converting image to grayscale
#It takes two arguments. one being the input image and the other being color space conversion code
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
imshow('Converting to grayscale image',gray_image)
```
