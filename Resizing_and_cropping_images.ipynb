```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title='Image',image=None,size=10):
  w=image.shape[0]
  h=image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
image=cv2.imread('/content/images/Hillary.jpg')   ##SCALING AN IMAGE
imshow('Scaling Image',image,size=6)
```

```python
image_scaled=cv2.resize(image,None,fx=0.75,fy=0.75)
imshow('scaling the image 0.75 times the original',image_scaled,size=6)
```

```python
image_scaled2=cv2.resize(image,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)
imshow('Scaling the image twice using inter cubic',image_scaled2,size=6)
```

```python
image_scaled3=cv2.resize(image,None,fx=2,fy=2,interpolation=cv2.INTER_NEAREST)
imshow('scaling the image twice using inter nearest',image_scaled3,size=6)
```

```python
#screwing the image by exact dimmensions
image_scaled4=cv2.resize(image,(900,400),interpolation=cv2.INTER_AREA)
imshow('scaling using inter area',image_scaled4,size=6)
```

```python
image=cv2.imread('/content/images/blacksandbeach.jpeg') ###IMAGE PYRAMIDS

smaller=cv2.pyrDown(image)
larger=cv2.pyrUp(smaller)

imshow('original',image,size=6)
imshow('smaller',smaller,size=6)
imshow('larger',larger,size=6)

even_smaller=cv2.pyrDown(smaller)
imshow('even smaller',even_smaller,size=6)
```

```python
image=cv2.imread('/content/images/abraham_mask.png')
height,width=image.shape[:2]
start_row=int(height*.25) ##getting the starting pixel coordinates
start_col=int(width*.25)
end_row=int(height*.75) ##getting the ending pixel coordinates
end_col=int(width*.75)

cropped=image[start_row:end_row,start_col:end_col]

imshow('original image',image)
copy=image.copy()
cv2.rectangle(copy,(start_col,start_row),(end_col,end_row),(0,255,255),10)
imshow('area we are cropping',copy)
imshow('cropped image',cropped)
```
