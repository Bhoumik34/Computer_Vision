```python
#loading an image with 'imread'
image=cv2.imread('/content/images/Sunflowers.jpg')
```

```python
import matplotlib.pyplot as plt
# 'imshow' is used for showing the image with matplotlib
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def imshow(title='',image=None):
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
imshow('Displaying our first image',image)
```

```python
#saving images using 'imwrite' in jpg or png formats
cv2.imwrite('output.jpg',image)
```

```python
cv2.imwrite('output.png',image)
```
